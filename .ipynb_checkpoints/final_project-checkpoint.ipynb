```python
a=list(cd['type'])
for i in range(0,len(cd)):
    if str(a[i][0])=='I' and a[i][1]=='n':
        a[i]="Industrail"
    elif str(a[i][0])=="R"and a[i][1]=='e':
        a[i]="Residential"
    else:
        a[i]="other"
        
cd['type'] = a
cd['type'].value_counts()
```

# Line Plot
Purpose: Visualize trends over time for different pollutants.
Example: Plot SO2, NO2, RSPM, and SPM levels over time to see how they change.

```python
plt.figure(figsize=(10, 6))
plt.plot(cd['date'], cd['so2'], label='SO2', marker='o')
plt.plot(cd['date'], cd['no2'], label='NO2', marker='o')
plt.plot(cd['date'], cd['rspm'], label='RSPM', marker='o')
plt.plot(cd['date'], cd['spm'], label='SPM', marker='o')
plt.xlabel('Date')
plt.ylabel('Concentration')
plt.title('Pollutant Levels Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
sns.catplot(x = "type", kind = "count", palette = "ch: 0.25", data = cd)
```

```python
# Select only numeric columns from your DataFrame
numeric_cd = cd.select_dtypes(include=[np.number])
# Now calculate the correlation matrix
corrmat = numeric_cd.corr()
# Plot the heatmap
f, ax = plt.subplots(figsize=(15, 10))
sns.heatmap(corrmat, vmax=1, square=True, annot=True, cmap='coolwarm', linewidths=0.5, cbar_kws={"shrink": 0.8})
plt.show()
```

```python
cd[['no2', 'location']].groupby(['location']).median().sort_values("no2", ascending = False).tail(50).plot.bar(color = 'g')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
cols = ['so2', 'no2', 'rspm', 'spm']
sns.pairplot(cd[cols + ['type']].dropna(), hue='type', height=2.5)
plt.show()
```

```python
total_levels = cd[['so2', 'no2', 'rspm', 'spm']].sum()
print(total_levels)
total_levels = {
    'SO2': 1.595577e+06,
    'NO2': 4.331986e+06,
    'RSPM': 1.704338e+07,
    'SPM': 3.507452e+07
}

# Create the pie chart
plt.figure(figsize=(10, 7))
plt.pie(total_levels.values(), labels=total_levels.keys(), autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title('Distribution of Total Pollutant Levels')
plt.show()
```

```python

# Calculate correlation matrix
correlation_matrix = cd[['so2', 'no2', 'rspm', 'spm']].corr()

print(correlation_matrix)
```

```python

# Group by state and calculate average levels
state_avg = cd.groupby('state')[['so2', 'no2', 'rspm', 'spm']].mean()

# Plot average levels by state
state_avg.plot(kind='bar', figsize=(14, 8))
plt.xlabel('State')
plt.ylabel('Average Concentration')
plt.title('Average Pollutant Levels by State')
plt.xticks(rotation=90)
plt.legend(title='Pollutants')
plt.show()
```

```python
# Create a box plot for pollutant levels
plt.figure(figsize=(12, 8))
cd[['so2', 'no2', 'rspm', 'spm']].boxplot(grid=False, patch_artist=True, 
                                          boxprops=dict(facecolor='lightblue', color='blue'),
                                          whiskerprops=dict(color='blue'),
                                          capprops=dict(color='blue'),
                                          medianprops=dict(color='red'))
plt.title('Box Plot of Pollutant Levels')
plt.ylabel('Concentration')
plt.xlabel('Pollutants')
plt.show()
```
