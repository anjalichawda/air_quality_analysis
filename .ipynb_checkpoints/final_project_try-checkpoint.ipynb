# Line Plot
Purpose: Visualize trends over time for different pollutants.
Example: Plot SO2, NO2, RSPM, and SPM levels over time to see how they change.

```python
plt.figure(figsize=(10, 6))
plt.plot(cd['date'], cd['so2'], label='SO2', marker='o')
plt.plot(cd['date'], cd['no2'], label='NO2', marker='o')
plt.plot(cd['date'], cd['rspm'], label='RSPM', marker='o')
plt.plot(cd['date'], cd['spm'], label='SPM', marker='o')
plt.xlabel('Date')
plt.ylabel('Concentration')
plt.title('Pollutant Levels Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
sns.catplot(x = "type", kind = "count", palette = "ch: 0.25", data = cd)
```

```python
# Select only numeric columns from your DataFrame
numeric_cd = cd.select_dtypes(include=[np.number])
# Now calculate the correlation matrix
corrmat = numeric_cd.corr()
# Plot the heatmap
f, ax = plt.subplots(figsize=(15, 10))
sns.heatmap(corrmat, vmax=1, square=True, annot=True, cmap='coolwarm', linewidths=0.5, cbar_kws={"shrink": 0.8})
plt.show()
```

# Step 1: Load the Data
# Replace 'data.csv' with the path to your dataset
# Step 2: Explore and Clean the Data
# Ensure 'rspm' and 'state' columns are present

```python
print(data[['rspm', 'state']].head())
```

```python
df = data
```

# Drop rows with missing 'rspm' or 'state'

```python
df_cleaned = df.dropna(subset=['rspm', 'state'])
```

# Convert 'rspm' to numeric, if not already

```python
df_cleaned['rspm'] = pd.to_numeric(df_cleaned['rspm'], errors='coerce')
```

```python
# Step 3: Aggregate Data by State
# Compute the average RSPM level for each state
state_rspm = df_cleaned.groupby('state')['rspm'].mean().reset_index()
```

```python
# Step 4: Plot RSPM Levels State-wise
plt.figure(figsize=(14, 8))
sns.barplot(x='state', y='rspm', data=state_rspm, palette='viridis')
plt.xlabel('State')
plt.ylabel('Average RSPM Level')
plt.title('Average RSPM Levels by State')
plt.xticks(rotation=90)  # Rotate state names for better readability
plt.grid(axis='y')
plt.show()
```

Plot RSPM levels state and location wise
# Step 1: Load the Data
# Replace 'data.csv' with the path to your dataset

```python
gf = data
```

```python
# Step 2: Explore and Clean the Data
# Ensure 'rspm', 'state', and 'location' columns are present
print(gf[['rspm', 'state', 'location']].head())
```

```python
# Drop rows with missing 'rspm', 'state', or 'location'
gf_cleaned = gf.dropna(subset=['rspm', 'state', 'location'])
```

```python
# Convert 'rspm' to numeric, if not already
gf_cleaned['rspm'] = pd.to_numeric(gf_cleaned['rspm'], errors='coerce')
```

```python
# Step 3: Aggregate Data by State and Location
# Compute the average RSPM level for each state and location combination
state_location_rspm = df_cleaned.groupby(['state', 'location'])['rspm'].mean().reset_index()
```

```python
# Step 4: Plot RSPM Levels State and Location-wise
plt.figure(figsize=(16, 10))
sns.barplot(x='location', y='rspm', hue='state', data=state_location_rspm, palette='viridis')
plt.xlabel('Location')
plt.ylabel('Average RSPM Level')
plt.title('Average RSPM Levels by State and Location')
plt.xticks(rotation=90)  # Rotate location names for better readability
plt.legend(title='State')
plt.grid(axis='y')
plt.show()
```
